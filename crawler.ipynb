```python
# rename image's name
import os
from PIL import Image

def compress_image(path):
    basewidth = 200
    img = Image.open(path)
    wpercent = (basewidth/float(img.size[0]))
    hsize = int((float(img.size[1])*float(wpercent)))
    img = img.resize((basewidth, hsize), Image.ANTIALIAS)
    try:
        img.save(path)
    except:
        print('resize error', path)

root_path = 'downloads'
dirs = os.listdir(root_path)
for dir_name in dirs:
    files = os.listdir(os.path.join(root_path, dir_name))
    for fid, file in enumerate(files):
        file_name = os.path.join(root_path, dir_name, file)
        print(file_name)
        #os.rename(file_name, os.path.join(root_path, dir_name, str(fid) + '.jpg'))
        compress_image(file_name)
```
